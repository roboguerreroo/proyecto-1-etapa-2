# Modulación AM de señales con múltiples componentes armónicos

```python
def func_mt(Am1,fm1,Am2,fm2,Am3,fm3,t):
   m1t=Am1*np.cos(2*math.pi*fm1*t) # Componente 1
   m2t=Am2*np.cos(2*math.pi*fm2*t) # Componente 2
   m3t=Am3*np.cos(2*math.pi*fm3*t) # Componente 3
   mt=m1t+m2t+m3t # Señal combinada
   return mt

def func_ct(Ac,fc,t):
   ct=Ac*np.cos(2*math.pi*fc*t) # Señal portadora
   return ct

def Am_mod(mt,ct,Ka):
   st=(1+Ka*mt)*ct # Modulación AM
   return st
```

```python
Am1=3 # Amplitud de la componente 1
fm1=1000 # Frecuencia de la componente 1

Am2=2 # Amplitud de la componente 2
fm2=2000 # Frecuencia de la componente 2

Am3=1 # Amplitud de la componente 3
fm3=3000 # Frecuencia de la componente 3

Ac=1 # Amplitud de la portadora
fc=10000 # Frecuencia de la portadora
Ka=0.3 # Coeficiente de sensibilidad de amplitud

fm1_3=[fm1,fm2,fm3]

## -- No es necesario modificar el código después de esta línea
## Cálculo de la frecuencia fudamental de la señal, para su graficación.
fm=np.gcd.reduce(fm1_3)

## ---- Dominio del tiempo
Tstart=0
Tstop=4/fm
Tstep=0.0001/fm

t=np.arange(Tstart,Tstop,Tstep) # Tiempo
mt=func_mt(Am1,fm1,Am2,fm2,Am3,fm3,t) # Señal mensaje
ct=func_ct(Ac,fc,t) # Portadora
st=Am_mod(mt,ct,Ka) # Modulación AM
At=np.abs(Ac*(1+Ka*mt))

f = plt.figure(figsize=(20, 15))
ax1 = f.add_subplot(621)
ax2 = f.add_subplot(622)
ax3 = f.add_subplot(623)
ax4 = f.add_subplot(624)
ax5 = f.add_subplot(625)
ax6 = f.add_subplot(626)
ax7 = f.add_subplot(627)
ax8 = f.add_subplot(628)
ax9 = f.add_subplot(6,2,9)
ax10 = f.add_subplot(6,2,10)
ax11 = f.add_subplot(6,2,11)
ax12 = f.add_subplot(6,2,12)

# Graficar las señales
ax1.plot(t,mt,'b',label='$m(t)$')
ax1.set_xlabel('time (sec)')
ax1.set_ylabel('$m(t)$ (v)')
ax1.legend()

ax3.plot(t,ct,'k',label='$c(t)$')
ax3.set_xlabel('time (sec)')
ax3.set_ylabel('$c(t)$ (v)')
ax3.legend()

ax5.plot(t,At,'b--')
ax5.plot(t,-1*At,'k--')
ax5.plot(t,st,'r',label='$s(t)$')
ax5.set_xlabel('time (sec)')
ax5.set_ylabel('$s(t)$ (v)')
ax5.legend()

## ---- Dominio de la frecuencia
Tstart=0; Tstop=100000/fm; Tstep=0.05/fm
t=np.arange(Tstart,Tstop,Tstep) # Tiempo
N=len(t)
mt=func_mt(Am1,fm1,Am2,fm2,Am3,fm3,t) # Mensaje

Mf = np.abs(np.fft.fft(mt))
freqs = np.fft.fftfreq(mt.size, Tstep)
idx = np.argsort(freqs)
freqs=freqs[idx]
Mf=Mf[idx]/N
ax2.plot(freqs, Mf,'b',label='$M(f)$')
ax2.set_xlabel('freq (Hz)')
ax2.set_ylabel('$S(f)$ (v/Hz)')
ax2.legend()

Tstart=0; Tstop=100000/fc; Tstep=0.25/fc
t=np.arange(Tstart,Tstop,Tstep) # Tiempo
N=len(t)
mt=func_mt(Am1,fm1,Am2,fm2,Am3,fm3,t) # Mensaje
ct=func_ct(Ac,fc,t) # Portadora
st=Am_mod(mt,ct,Ka) # Modulación AM
At=np.abs(Ac*(1+Ka*mt))

Cf = np.abs(np.fft.fft(ct))
freqs = np.fft.fftfreq(ct.size, Tstep)
idx = np.argsort(freqs)
freqs=freqs[idx]
Cf=Cf[idx]/N
ax4.plot(freqs, Cf,'k',label='$C(f)$')
ax4.set_xlabel('frecuencia (Hz)')
ax4.set_ylabel('$C(f)$ (v/Hz)')
ax4.legend()

Sf = np.abs(np.fft.fft(st))
freqs = np.fft.fftfreq(st.size, Tstep)
idx = np.argsort(freqs)
freqs=freqs[idx]
Sf=Sf[idx]/N
ax6.plot(freqs, Sf,'r',label='$S(f)$')
ax6.set_xlabel('frecuencia (Hz)')
ax6.set_ylabel('$S(f)$ (v/Hz)')
ax6.legend()

# modificaciones en el dominio del tiempo

Am1=30 # Amplitud de la componente 1
fm1=10000 # Frecuencia de la componente 1

Am2=20 # Amplitud de la componente 2
fm2=20000 # Frecuencia de la componente 2

Am3=10 # Amplitud de la componente 3
fm3=30000 # Frecuencia de la componente 3

Ac=1 # Amplitud de la portadora
fc=100000 # Frecuencia de la portadora
Ka=0.3 # Coeficiente de sensibilidad de amplitud

fm1_3=[fm1,fm2,fm3]

fm=np.gcd.reduce(fm1_3)

Tstart=0
Tstop=4/fm
Tstep=0.0001/fm

t=np.arange(Tstart,Tstop,Tstep) # Tiempo
mt=func_mt(Am1,fm1,Am2,fm2,Am3,fm3,t) # Señal mensaje
ct=func_ct(Ac,fc,t) # Portadora
st=Am_mod(mt,ct,Ka) # Modulación AM
At=np.abs(Ac*(1+Ka*mt))

ax7.plot(t,mt,'b',label='$m(t)$')
ax7.set_title('Modificando fm1, fm2, fm3')
ax7.set_xlabel('time (sec)')
ax7.set_ylabel('$m(t)$ (v)')
ax7.legend()

ax9.plot(t,ct,'k',label='$c(t)$')
ax9.set_title('Modificando fc')
ax9.set_xlabel('time (sec)')
ax9.set_ylabel('$c(t)$ (v)')
ax9.legend()

ax11.plot(t,At,'b--')
ax11.plot(t,-1*At,'k--')
ax11.plot(t,st,'r',label='$s(t)$')
ax11.set_title('Modificando Am1, Am2, Am3')
ax11.set_xlabel('time (sec)')
ax11.set_ylabel('$s(t)$ (v)')
ax11.legend()

# modificaciones en el dominio de la frecuencia

Tstart=0; Tstop=100000/fm; Tstep=0.05/fm
t=np.arange(Tstart,Tstop,Tstep) # Tiempo
N=len(t)
mt=func_mt(Am1,fm1,Am2,fm2,Am3,fm3,t) # Mensaje

Mf = np.abs(np.fft.fft(mt))
freqs = np.fft.fftfreq(mt.size, Tstep)
idx = np.argsort(freqs)
freqs=freqs[idx]
Mf=Mf[idx]/N
ax8.plot(freqs, Mf,'b',label='$M(f)$')
ax8.set_title('Modificando fm1, fm2, fm3')
ax8.set_xlabel('freq (Hz)')
ax8.set_ylabel('$S(f)$ (v/Hz)')
ax8.legend()

Tstart=0; Tstop=100000/fc; Tstep=0.25/fc
t=np.arange(Tstart,Tstop,Tstep) # Tiempo
N=len(t)
mt=func_mt(Am1,fm1,Am2,fm2,Am3,fm3,t) # Mensaje
ct=func_ct(Ac,fc,t) # Portadora
st=Am_mod(mt,ct,Ka) # Modulación AM
At=np.abs(Ac*(1+Ka*mt))

Cf = np.abs(np.fft.fft(ct))
freqs = np.fft.fftfreq(ct.size, Tstep)
idx = np.argsort(freqs)
freqs=freqs[idx]
Cf=Cf[idx]/N
ax10.plot(freqs, Cf,'k',label='$C(f)$')
ax10.set_title('Modificando fc')
ax10.set_xlabel('frecuencia (Hz)')
ax10.set_ylabel('$C(f)$ (v/Hz)')
ax10.legend()

Sf = np.abs(np.fft.fft(st))
freqs = np.fft.fftfreq(st.size, Tstep)
idx = np.argsort(freqs)
freqs=freqs[idx]
Sf=Sf[idx]/N
ax12.plot(freqs, Sf,'r',label='$S(f)$')
ax12.set_title('Modificando Am1, Am2, Am3')
ax12.set_xlabel('frecuencia (Hz)')
ax12.set_ylabel('$S(f)$ (v/Hz)')
ax12.legend()

plt.subplots_adjust(hspace=1)
plt.show()
```
